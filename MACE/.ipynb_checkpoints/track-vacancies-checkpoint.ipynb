# Load libraries and set device up

```python
import matplotlib.pyplot as plt
import numpy             as np
import library.dynamics  as dyn
import json

from scipy.signal     import find_peaks
```

# Define parameters

```python
conf_path = "/Users/cibran/Desktop/ss_0.71875"
md_path   = f"{conf_path}/conf3/NPT/1500K"
```

```python
step_skip  = 2
step_equiv = 100
```

```python
# RDF parameters
r_max  = 5  # maximum distance in Å
dr     = 0.05  # bin width in Å
r_bins = np.arange(0, r_max + dr, dr)
r_axis = r_bins[:-1] + dr/2
```

# Perform analysis

```python
md_info, stc_info, centers, diffusion = dyn.md_analysis(
    md_path=md_path,
    ref_path=conf_path,
    step_skip=step_skip,
    step_equiv=step_equiv
)
```

```python
# Run the extraction
vacancies_t, interstitials_t, diffusion_t = dyn.get_defect_evolution(md_info, stc_info)
md_steps = len(vacancies_t)

print(f"Time steps: {len(vacancies_t)}")
print(f"Vacancies (first 10 steps):     {vacancies_t[:10]}")
print(f"Interstitials (first 10 steps): {interstitials_t[:10]}")
print(f"Diffusing (first 10 steps):     {diffusion_t[:10]}")

# Read time data
equiv_time, prod_time, traj_timestep, _, _ = dyn.load_trajectory(md_path, step_skip, step_equiv, data_format='raw')
```

```python
# Load simulation parameters
with open(f'{md_path}/npt-convergence.json', 'r') as f:
    params = json.load(f)
md_volume = params['volume (A^3)']['mean']

vac_counts = []
int_counts = []
vac_rdfs = 0
int_rdfs = 0
for t in range(md_steps):
    # Vacancies at t
    vacancy_positions = [pos['fractional_coordinates'] for pos in stc_info.values() if pos['when'][t]==0]

    # Interstitials at t
    interstitial_positions = []
    for atom in md_info.values():
        for center in atom['interstitial'].values():
            if center['when'][t] == 1:
                interstitial_positions.append(center['fractional_coordinates'])

    # Append
    vac_counts.append(len(vacancy_positions))
    int_counts.append(len(interstitial_positions))

    vac_rdf = dyn.compute_rdf(vacancy_positions,      md_volume, r_max, dr)
    int_rdf = dyn.compute_rdf(interstitial_positions, md_volume, r_max, dr)

    # Add to rdfs
    vac_rdfs += vac_rdf
    int_rdfs += int_rdf

# Average
vac_rdfs /= md_steps
int_rdfs /= md_steps
```

```python
# Save vacancy counts
np.savetxt(f"{md_path}/vacancy_counts.txt", vac_counts)
np.savetxt(f"{md_path}/vacancy_rdf.txt",    vac_rdfs)
np.savetxt(f"{md_path}/interstitial_counts.txt", int_counts)
np.savetxt(f"{md_path}/interstitial_rdf.txt",    int_rdfs)
```

```python
time_ps = np.arange(md_steps) * traj_timestep * 1e-3  # ps

plt.plot(time_ps, vac_counts, '-o')
plt.xlabel("t (ps)")
plt.ylabel(r"$n_{\mathrm{vac}}$")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(f"{md_path}/vacancy_evolution.pdf", dpi=30)
plt.show()
```

```python
plt.plot(time_ps, int_counts, '-o')
plt.xlabel("t (ps)")
plt.ylabel(r"$n_{\mathrm{int}}$")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(f"{md_path}/interstitial_evolution.pdf", dpi=30)
plt.show()
```

```python
plt.plot(r_axis, vac_rdfs, ':o')
plt.xlabel("r (Å)")
plt.ylabel(r"$g_{\mathrm{vac}}(r)$")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(f"{md_path}/vacancy_rdf.pdf", dpi=30)
plt.show()
```

```python
plt.plot(r_axis, int_rdfs, ':o')
plt.xlabel("r (Å)")
plt.ylabel(r"$g_{\mathrm{int}}(r)$")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(f"{md_path}/interstitial_rdf.pdf", dpi=30)
plt.show()
```

# Load and gather data

```python
data = {}
for x in ['0.46875', '0.71875', '0.84375', '0.96875', '0.53125', '0.75', '0.875', '0.59375', '0.78125', '0.90625', '0.4375', '0.65625', '0.8125', '0.9375']:
    data[x] = {}
    for temp in ['1200K', '1500K', '1700K', '2000K']:
        data[x][temp] = {}
        for conf in ['conf1', 'conf2', 'conf3']:
            # Load RDF
            conf_path = f"/home/cibran/Desktop/GdCeO2-MLMD-production/ss_{x}/{conf}/NPT/{temp}/reduced"
            rdf            = np.loadtxt(f'{conf_path}/vacancy_rdf.txt')
            vacancy_counts = np.loadtxt(f'{conf_path}/vacancy_counts.txt')
            
            # Find peaks
            peaks, _  = find_peaks(rdf, prominence=0.5)
            r_peaks   = r_axis[peaks]
            rdf_peaks = rdf[peaks]
            
            # Store in dictionary
            data[x][temp][conf] = {
                'r_peaks':   r_peaks.tolist(),
                'rdf_peaks': rdf_peaks.tolist(),
                'vacancy_counts_mean': np.mean(vacancy_counts),
                'vacancy_counts_std': np.std(vacancy_counts)
            }
            
            #plt.plot(r_axis, rdf, label=temp)

        #plt.xlabel("r (Å)")
        #plt.ylabel(r"$g_{\mathrm{vac}}(r)$")
        #plt.grid(alpha=0.3)
        #plt.xlim(left=2)
        #plt.legend(loc='best')
        #plt.tight_layout()
        #plt.savefig(f"{conf_path}/vacancy_rdfs.pdf", dpi=30)
        #plt.show()
```

```python
# --- Get x values, temperatures, and configurations ---
x_values = sorted([float(x) for x in data.keys()])
first_x = str(x_values[0])
temps = list(data[first_x].keys())
confs = list(data[first_x][temps[0]].keys())  # configurations under a temperature

# --- Compute global min and max of peak heights for consistent colorbar ---
all_heights_temp = []
for x in data:
    for conf in data[x][temp]:
        all_heights_temp.extend(data[x][temp][conf]['rdf_peaks'])
vmin = 0
vmax = np.ceil(np.nanmax(all_heights_temp))

# --- Plot per temperature ---
for temp in temps:    
    # Find maximum number of peaks across all concentrations for this temperature
    max_peaks = 0
    for x in x_values:
        for conf in confs:
            max_peaks = max(max_peaks, len(data[str(x)][temp][conf]['r_peaks']))
    
    # Prepare storage for each peak index
    peak_positions_per_index = [[] for _ in range(max_peaks)]
    peak_std_per_index       = [[] for _ in range(max_peaks)]
    peak_heights_per_index   = [[] for _ in range(max_peaks)]
    
    # --- Compute mean, std, and heights across configurations ---
    for x in x_values:
        r_peaks_list = []
        rdf_peaks_list = []
        for conf in confs:
            r_peaks_list.append(np.array(data[str(x)][temp][conf]['r_peaks']))
            rdf_peaks_list.append(np.array(data[str(x)][temp][conf]['rdf_peaks']))
        
        # Pad arrays with NaN for variable number of peaks
        n_peaks = max(len(p) for p in r_peaks_list)
        r_arr = np.full((len(confs), n_peaks), np.nan)
        h_arr = np.full((len(confs), n_peaks), np.nan)
        for i in range(len(confs)):
            r_arr[i, :len(r_peaks_list[i])] = r_peaks_list[i]
            h_arr[i, :len(rdf_peaks_list[i])] = rdf_peaks_list[i]
        
        # Compute mean and std across configurations
        mean_r = np.nanmean(r_arr, axis=0)
        std_r  = np.nanstd(r_arr, axis=0)
        mean_h = np.nanmean(h_arr, axis=0)
        
        # Store values per peak index
        for i in range(n_peaks):
            peak_positions_per_index[i].append(mean_r[i])
            peak_std_per_index[i].append(std_r[i])
            peak_heights_per_index[i].append(mean_h[i])
        
        # Pad remaining peak indices
        for i in range(n_peaks, max_peaks):
            peak_positions_per_index[i].append(np.nan)
            peak_std_per_index[i].append(np.nan)
            peak_heights_per_index[i].append(np.nan)
    
    # --- Plot peaks with lines, error bars, and scatter with global colorbar ---
    for i in range(max_peaks):
        y = np.array(peak_positions_per_index[i])
        yerr = np.array(peak_std_per_index[i])
        h = np.array(peak_heights_per_index[i])
        
        # Connect peaks
        plt.plot(x_values, y, linestyle='-', color='gray')
        
        # Error bars
        plt.errorbar(x_values, y, yerr=yerr, fmt='none')
        
        # Scatter with color-coded heights
        sc = plt.scatter(x_values, y, c=h, cmap='viridis', vmin=vmin, vmax=vmax)
    
    plt.xlabel(r'$\mathregular{Ce_{x}Gd_{1-x}O_{2-x/2}}$')
    plt.ylabel("r (Å)")
    plt.title(f'T = {temp}')
    cbar = plt.colorbar(sc)
    cbar.set_label(r"$g_{\mathrm{vac}}(r)$")
    plt.savefig(f'results-T={temp}.pdf', dpi=50, bbox_inches='tight')
    plt.show()
```

```python
# --- Get x values, temperatures, and configurations ---
x_values = sorted([float(x) for x in data.keys()])
first_x = str(x_values[0])
temps = list(data[first_x].keys())
confs = list(data[first_x][temps[0]].keys())  # configurations under a temperature

# --- Find maximum number of peaks across all concentrations and temperatures ---
max_peaks = 0
for x in x_values:
    for temp in temps:
        for conf in confs:
            max_peaks = max(max_peaks, len(data[str(x)][temp][conf]['r_peaks']))

# --- Compute mean r for each peak index for the title ---
peak_positions_avg = []
for peak_idx in range(max_peaks):
    r_all = []
    for x in x_values:
        for temp in temps:
            for conf in confs:
                r_peaks = data[str(x)][temp][conf]['r_peaks']
                if peak_idx < len(r_peaks):
                    r_all.append(r_peaks[peak_idx])
    if r_all:
        peak_positions_avg.append(np.mean(r_all))
    else:
        peak_positions_avg.append(np.nan)

# --- Loop over each peak index ---
for peak_idx in range(max_peaks):        
    # --- Loop over temperatures ---
    for temp in temps:
        y_vals = []
        y_errs = []
        
        for x in x_values:
            r_peaks_list = []
            rdf_peaks_list = []
            for conf in confs:
                r_peaks_list.append(np.array(data[str(x)][temp][conf]['r_peaks']))
                rdf_peaks_list.append(np.array(data[str(x)][temp][conf]['rdf_peaks']))
            
            # Pad arrays with NaN to handle variable number of peaks
            n_peaks = max(len(p) for p in r_peaks_list)
            h_arr = np.full((len(confs), n_peaks), np.nan)
            for i in range(len(confs)):
                h_arr[i, :len(rdf_peaks_list[i])] = rdf_peaks_list[i]
            
            # Mean and std across configurations
            mean_h = np.nanmean(h_arr, axis=0)
            std_h  = np.nanstd(h_arr, axis=0)
            
            # For this peak index, check if it exists
            if peak_idx < len(mean_h):
                y_vals.append(mean_h[peak_idx])
                y_errs.append(std_h[peak_idx])
            else:
                y_vals.append(np.nan)
                y_errs.append(np.nan)
        
        # Convert to numpy arrays
        y_vals = np.array(y_vals)
        y_errs = np.array(y_errs)
        
        # Plot line connecting this temperature
        plt.plot(x_values, y_vals, label=temp, linestyle='-', marker='o')
        
        # Error bars
        plt.errorbar(x_values, y_vals, yerr=y_errs, fmt='none', ecolor='gray')
    
    plt.xlabel(r'$\mathregular{Ce_{x}Gd_{1-x}O_{2-x/2}}$')
    plt.ylabel("$g_{\mathrm{vac}}(r)$")
    
    # Use average r position in the title
    avg_r = peak_positions_avg[peak_idx]
    if not np.isnan(avg_r):
        plt.title(f'Peak at r ≈ {avg_r:.2f} Å')
    else:
        plt.title(f'Peak index {peak_idx + 1}')
    
    plt.legend(title='Temperature')
    plt.savefig(f'results-peak={peak_idx}.pdf', dpi=50, bbox_inches='tight')
    plt.show()
```

# Plot with temperature

```python
initial_vacancies = {
    '0.4375':  108,
    '0.46875': 102,
    '0.53125': 90,
    '0.59375': 78,
    '0.65625': 66,
    '0.71875': 54,
    '0.75':    48,
    '0.78125': 42,
    '0.8125':  36,
    '0.84375': 30,
    '0.875':   24,
    '0.90625': 18,
    '0.9375':  12,
    '0.96875': 6,
}
```

```python
# --- Get x values, temperatures, and configurations ---
x_values = sorted([float(x) for x in data.keys()])
first_x = str(x_values[0])
temps = list(data[first_x].keys())
confs = list(data[first_x][temps[0]].keys())

# --- Prepare plot ---
plt.figure(figsize=(10, 6))

for temp in temps:
    mean_vacancies = []
    std_vacancies  = []

    for x in x_values:
        mu_list = []
        sigma_list = []
        for conf in confs:
            mu_list.append(data[str(x)][temp][conf]['vacancy_counts_mean'])
            sigma_list.append(data[str(x)][temp][conf]['vacancy_counts_std'])
        
        mu_arr = np.array(mu_list)
        sigma_arr = np.array(sigma_list)
        
        # Overall mean
        mean_total = np.mean(mu_arr)
        
        # Propagated std: sqrt(Var_between + mean(Var_within))
        var_between = np.var(mu_arr, ddof=0)  # variance of the means
        var_within  = np.mean(sigma_arr**2)  # average of individual variances
        std_total   = np.sqrt(var_between + var_within)
        
        mean_vacancies.append(mean_total)
        std_vacancies.append(std_total)
    
    mean_vacancies = np.array(mean_vacancies)
    std_vacancies  = np.array(std_vacancies)
    
    # Plot line + error bars
    plt.plot(x_values, mean_vacancies, marker='o', linestyle='-', label=temp)
    plt.errorbar(x_values, mean_vacancies, yerr=std_vacancies, fmt='none', ecolor='gray', alpha=0.6)

original_vacancies = [initial_vacancies[str(x)] for x in x_values]
plt.plot(x_values, original_vacancies, marker='o', linestyle='-', label='0K')

plt.xlabel(r'$\mathregular{Ce_{x}Gd_{1-x}O_{2-x/2}}$')
plt.ylabel("Number of vacancies")
plt.legend(title="Temperature")
plt.savefig(f'results-vacancies.pdf', dpi=50, bbox_inches='tight')
plt.show()
```
